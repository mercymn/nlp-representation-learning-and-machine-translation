```python
# DATA EXPLORATION BEFORE PREPROCESSING
# 1. Dataset Size
num_docs_small = len(wikitext_small)
num_docs_large = len(wikitext_large)

print("DATASET SIZE:")
print(f"  Documents (small): {num_docs_small:,}")
print(f"  Documents (large): {num_docs_large:,}")
print()

# 2. Raw token counts
total_tokens_small = sum(len(doc.split()) for doc in wikitext_small['text'])
total_tokens_large = sum(len(doc.split()) for doc in wikitext_large['text'])

print("TOKEN COUNTS:")
print(f"  Total tokens (small): {total_tokens_small:,}")
print(f"  Total tokens (large): {total_tokens_large:,}")
print(f"  Avg doc length (small): {total_tokens_small/num_docs_small:.1f} tokens")
print(f"  Avg doc length (large): {total_tokens_large/num_docs_large:.1f} tokens")
print()

# 3. Raw vocabulary size
raw_vocab_small = set(word.lower() for doc in wikitext_small['text'] for word in doc.split())
raw_vocab_large = set(word.lower() for doc in wikitext_large['text'] for word in doc.split())

print("RAW VOCABULARY SIZE:")
print(f"  Vocabulary (small): {len(raw_vocab_small):,}")
print(f"  Vocabulary (large): {len(raw_vocab_large):,}")
print()

# 4. Sample documents
print("SAMPLE DOCUMENTS:")
print(f"  Small corpus [0]: {wikitext_small['text'][0]}")
print(f"  Large corpus [0]: {wikitext_large['text'][0]}")
print()

# 5. Data quality checks
empty_small = sum(1 for doc in wikitext_small['text'] if len(doc.strip()) == 0)
empty_large = sum(1 for doc in wikitext_large['text'] if len(doc.strip()) == 0)

print("DATA QUALITY:")
print(f"  Empty documents (small): {empty_small}")
print(f"  Empty documents (large): {empty_large}")
print()

# 6. Stopword analysis
from gensim.parsing.preprocessing import STOPWORDS

# Number of stopwords in vocabularies
stopwords_in_vocab_small = [w for w in raw_vocab_small if w in STOPWORDS]
stopwords_in_vocab_large = [w for w in raw_vocab_large if w in STOPWORDS]

print("STOPWORDS IN VOCABULARY:")
print(f"  Stopwords (small): {len(stopwords_in_vocab_small)} out of {len(STOPWORDS)}")
print(f"  First 10: {stopwords_in_vocab_small[:10]}")
print()
print(f"  Stopwords (large): {len(stopwords_in_vocab_large)} out of {len(STOPWORDS)}")
print(f"  First 10: {stopwords_in_vocab_large[:10]}")
print()

# Missing stopwords
all_stopwords = set(STOPWORDS)

stopwords_small_set = set(stopwords_in_vocab_small)
stopwords_large_set = set(stopwords_in_vocab_large)

missing_small = all_stopwords - stopwords_small_set
missing_large = all_stopwords - stopwords_large_set

print(f"Missing from small corpus: {sorted(missing_small)}")
print(f"Missing from large corpus: {sorted(missing_large)}")
```

```python
# Step 2
# References:
# Mikolov, T., Chen, K., Corrado, G., & Dean, J. (2013).
#   Efficient Estimation of Word Representations in Vector Space.
#   arXiv preprint arXiv:1301.3781.
#  https://arxiv.org/abs/1301.3781

# Gensim Documentation – Word2Vec and Preprocessing Utilities:
#   https://radimrehurek.com/gensim/models/word2vec.html
#   https://radimrehurek.com/gensim/utils.html
#
# NLTK Documentation – WordNet Lemmatizer:
#   https://www.nltk.org/api/nltk.stem.wordnet.html

# The preprocessing uses:
#   - simple_preprocess (Gensim)
#   - remove_stopwords (Gensim)
#   - WordNetLemmatizer (NLTK)
#   for text cleaning and normalization before training Word2Vec models.

from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import remove_stopwords
from nltk.stem import WordNetLemmatizer
from gensim.models import Word2Vec
import nltk

# Required NLTK data download
nltk.download('wordnet')
nltk.download('omw-1.4')

# ---------------------------------------------------------
# Preprocessing function with lemmatization
# ---------------------------------------------------------
def preprocess_texts(dataset, use_lemmatization=True):
    """
    Cleans and tokenizes the dataset using Gensim
    - Removes stopwords
    - Tokenizes words with simple_preprocess
    - Lemmatizes using NLTK
    Returns: list of token lists
    """
    lemmatizer = WordNetLemmatizer() if use_lemmatization else None
    processed = []

    for text in dataset['text']:
        if isinstance(text, str) and text.strip():
            # Remove stopwords
            cleaned = remove_stopwords(text.lower())
            # Tokenize using gensim's simple_preprocess
            tokens = simple_preprocess(cleaned, deacc=True)

            # Apply lemmatization if enabled
            if use_lemmatization:
                tokens = [lemmatizer.lemmatize(token) for token in tokens]

            processed.append(tokens)

    return processed

# ---------------------------------------------------------
# Preprocess both corpora
# ---------------------------------------------------------
train_small = preprocess_texts(wikitext_small, use_lemmatization=True)
train_large = preprocess_texts(wikitext_large, use_lemmatization=True)

print(f"Sample tokens (small): {train_small[0][:10]}")
print(f"Sample tokens (large): {train_large[0][:10]}")

# ---------------------------------------------------------
# Define baseline parameters
# ---------------------------------------------------------
params = dict(vector_size=50, window=5, min_count=5, epochs=5)

# ---------------------------------------------------------
# Train models
# ---------------------------------------------------------
print("Training Word2Vec on wikitext_small...")
model_small = Word2Vec(sentences=train_small, **params)
print("Small model trained.")

print("Training Word2Vec on wikitext_large...")
model_large = Word2Vec(sentences=train_large, **params)
print("Large model trained.")
```

```python
# Step 3: Semantic Relatedness Task
# Reference:
# References:
# Agirre, E., Alfonseca, E., Hall, K., Kravalova, J., Paşca, M., & Soroa, A. (2009).
#   A Study on Similarity and Relatedness Using Distributional and WordNet-based Approaches.
#   In Proceedings of NAACL-HLT 2009, Boulder, Colorado.
#   https://gabrilovich.com/resources/data/wordsim353/wordsim353.zip
#
# Mikolov, T., Chen, K., Corrado, G., & Dean, J. (2013).
#   Efficient Estimation of Word Representations in Vector Space.
#   arXiv preprint arXiv:1301.3781.
#
# Gensim Documentation – Word2Vec and Similarity Computations:
#   https://radimrehurek.com/gensim/models/keyedvectors.html
#
# SciPy Documentation – Cosine Distance Function:
#   https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.cosine.html
# ------------------------------------------------------------

import requests, zipfile, io, pandas as pd
from scipy.spatial.distance import cosine

# ------------------------------------------------------------
# 1. Download and extract the WordSim-353 dataset
# ------------------------------------------------------------
url = "https://gabrilovich.com/resources/data/wordsim353/wordsim353.zip"
response = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(response.content))
z.extractall("wordsim353")
print(" WordSim-353 downloaded and extracted successfully.")

# ------------------------------------------------------------
# 2. Load the 'combined.csv' file into a DataFrame
# ------------------------------------------------------------
ws353 = pd.read_csv("wordsim353/combined.csv")

# ------------------------------------------------------------
# 3. Define cosine similarity function (Gensim-compatible)
# ------------------------------------------------------------
def cosine_similarity(word1, word2, model):
    """
    Compute cosine similarity between two words.
    Works for both Word2Vec (trained) and KeyedVectors (pretrained) models.
    """
    # Determine key set depending on model type
    if hasattr(model, 'wv'):
        key_index = model.wv.key_to_index
        get_vec = model.wv.__getitem__
    else:
        key_index = model.key_to_index
        get_vec = model.__getitem__

    if word1 in key_index and word2 in key_index:
        return 1 - cosine(get_vec(word1), get_vec(word2))
    else:
        return None

# ------------------------------------------------------------
# 4. Compute similarities for required pairs
# ------------------------------------------------------------
pairs = [
    ("plane", "car"),
    ("planet", "sun"),
    ("cup", "article"),
    ("sugar", "approach")
]

print("\n=== Cosine Similarities (wikitext_small) ===")
for w1, w2 in pairs:
    sim = cosine_similarity(w1, w2, model_small)
    print(f"{w1:>10s} / {w2:<10s} : {sim}")

print("\n=== Cosine Similarities (wikitext_large) ===")
for w1, w2 in pairs:
    sim = cosine_similarity(w1, w2, model_large)
    print(f"{w1:>10s} / {w2:<10s} : {sim}")
```

```python
# ============================================================
# STEP 4 – Evaluate Semantic Relatedness (Full 353 pairs)
# ============================================================
# References:
# Agirre et al., 2009 – WordSim-353 Dataset
# https://gabrilovich.com/resources/data/wordsim353/wordsim353.zip
#
# Mikolov et al., 2013 – Efficient Estimation of Word Representations in Vector Space
# https://arxiv.org/abs/1301.3781
#
# SciPy Documentation – Spearman Rank-Order Correlation
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html
#
# Gensim Documentation – KeyedVectors & similarity computations
# https://radimrehurek.com/gensim/models/keyedvectors.html
# ============================================================

from scipy.spatial.distance import cosine
from scipy.stats import spearmanr
import pandas as pd
import numpy as np

# ------------------------------------------------------------
# 1. Compute cosine similarities for *all* 353 word pairs
# ------------------------------------------------------------
def compute_cosine_similarities(model, dataset):
    """
    Computes cosine similarity for every WordSim-353 pair.
    Returns a new DataFrame with a 'cosine_similarity' column.
    Out-of-vocabulary (OOV) pairs get similarity = 0
    to ensure 353 values per model.
    """
    similarities = []
    # choose correct vector accessor
    if hasattr(model, "wv"):
        key_index = model.wv.key_to_index
        get_vec = model.wv.__getitem__
    else:
        key_index = model.key_to_index
        get_vec = model.__getitem__

    for _, row in dataset.iterrows():
        w1, w2 = row["Word 1"].lower(), row["Word 2"].lower()
        if w1 in key_index and w2 in key_index:
            sim = 1 - cosine(get_vec(w1), get_vec(w2))
        else:
            sim = 0.0        # ensure 353 numbers
        similarities.append(sim)

    dataset = dataset.copy()
    dataset["cosine_similarity"] = similarities
    return dataset

# ------------------------------------------------------------
# 2. Evaluate Spearman correlation (full 353 pairs)
# ------------------------------------------------------------
def evaluate_spearman_full(model, dataset, label="Model"):
    benchmark_df = compute_cosine_similarities(model, dataset)

    # drop NaNs if any numerical issues arise
    valid_pairs = benchmark_df.dropna(subset=["cosine_similarity"])

    spearman_corr, spearman_pval = spearmanr(
        valid_pairs["Human (mean)"],
        valid_pairs["cosine_similarity"]
    )

    print(f"=== {label}: Spearman rank-order correlation ===")
    print(f"ρ = {spearman_corr:.4f}")
    print(f"p-value = {spearman_pval:.4e}")
    print(f"Valid pairs used: {len(valid_pairs)} / {len(dataset)}\n")

    return benchmark_df, spearman_corr, spearman_pval

# ------------------------------------------------------------
# 3. Run evaluation for both models
# ------------------------------------------------------------
benchmark_small, rho_small, p_small = evaluate_spearman_full(model_small, ws353, "wikitext_small")
benchmark_large, rho_large, p_large = evaluate_spearman_full(model_large, ws353, "wikitext_large")

# ------------------------------------------------------------
# 4. Show example subset of key pairs
# ------------------------------------------------------------
word_pairs = [
    ('plane', 'car'),
    ('planet', 'sun'),
    ('cup', 'article'),
    ('sugar', 'approach')
]

filtered_df = ws353[
    ws353.apply(
        lambda row: (row["Word 1"], row["Word 2"]) in word_pairs
                    or (row["Word 2"], row["Word 1"]) in word_pairs,
        axis=1
    )
]

print("Example Word Pair Results (from large model):")
print("-" * 50)
for _, row in filtered_df.iterrows():
    w1, w2 = row["Word 1"], row["Word 2"]
    human = row["Human (mean)"]
    cosine_val = compute_cosine_similarities(model_large, pd.DataFrame([row]))["cosine_similarity"].iloc[0]
    print(f"{w1:>10s} – {w2:<10s} | Human: {human:>5.2f} | Cosine: {cosine_val:>6.3f}")
```

```python
# ============================================================
# STEP 5: Evaluate Pretrained Word2Vec (Google News)
# ============================================================
# Reference:
# Mikolov et al., 2013 – Efficient Estimation of Word Representations in Vector Space
# Gensim Downloader Documentation:
#   https://radimrehurek.com/gensim/downloader.html
# ============================================================

import gensim.downloader as api
from scipy.spatial.distance import cosine
from scipy.stats import spearmanr
import numpy as np

# ------------------------------------------------------------
# 1. Load pretrained Word2Vec model (Google News)
# ------------------------------------------------------------
print("Loading pretrained 'word2vec-google-news-300' model (~1.6GB)...")
model_pretrained = api.load("word2vec-google-news-300")
print(f"Model loaded successfully! Vocabulary size: {len(model_pretrained)}")
print(f"Vector dimensions: {model_pretrained.vector_size}")

# ------------------------------------------------------------
# 2. Evaluate model on WordSim-353 (ensuring 353 pairs)
# ------------------------------------------------------------
def evaluate_model_on_wordsim(model, dataset):
    """
    Evaluate a pretrained Word2Vec model on WordSim-353.
    Computes cosine similarity for *all* 353 pairs.
    If a word is OOV (not in vocabulary), assigns similarity = 0.
    Returns Spearman correlation (ρ) and p-value.
    """
    predicted_sims, human_scores = [], []

    for _, row in dataset.iterrows():
        w1, w2 = row['Word 1'].lower(), row['Word 2'].lower()

        if w1 in model.key_to_index and w2 in model.key_to_index:
            sim = 1 - cosine(model[w1], model[w2])
        else:
            sim = 0.0  # handle out-of-vocabulary cases

        predicted_sims.append(sim)
        human_scores.append(row['Human (mean)'])

    # Compute Spearman correlation across all 353 pairs
    rho, p_value = spearmanr(predicted_sims, human_scores)
    return rho, p_value, len(predicted_sims)

# ------------------------------------------------------------
# 3. Run evaluation
# ------------------------------------------------------------
rho_pretrained, p_pretrained, n_pairs = evaluate_model_on_wordsim(model_pretrained, ws353)

print("\n" + "=" * 60)
print("STEP 5: PRETRAINED WORD2VEC EVALUATION")
print("=" * 60)
print(f"Spearman correlation (Google News pretrained): ρ = {rho_pretrained:.4f}")
print(f"P-value: {p_pretrained:.4e}")
print(f"Valid pairs: {n_pairs}/353")  # should now always show 353/353
print("=" * 60)
```

```python
# ---------------------------------------------------------
# Step 6: Hyperparameter tuning experiments
# ---------------------------------------------------------

# References:
# Mikolov, T., Chen, K., Corrado, G., & Dean, J. (2013).
#   Efficient Estimation of Word Representations in Vector Space.
#   arXiv preprint arXiv:1301.3781.  (Original Word2Vec paper)

# Merity, S., Xiong, C., Bradbury, J., & Socher, R. (2016).
#   Pointer Sentinel Mixture Models. arXiv:1609.07843.
#   (Example table structure used for reporting results)

# Levy, O., Goldberg, Y., & Dagan, I. (2015).
#   Improving Distributional Similarity with Lessons Learned from Word Embeddings.
#   Transactions of the Association for Computational Linguistics, 3, 211–225.
#   (Discussion of parameter effects like vector size, window, and corpus)

# Gensim Documentation:
#   https://radimrehurek.com/gensim/models/word2vec.html
import time
from gensim.models import Word2Vec
import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine
from scipy.stats import spearmanr

# Computes Spearman correlation both for valid-only pairs and all 353 pairs

def evaluate_model_on_wordsim_dual(model, dataset):
    if hasattr(model, "wv"):
        key_index = model.wv.key_to_index
        get_vec = model.wv.__getitem__
    else:
        key_index = model.key_to_index
        get_vec = model.__getitem__

    predicted_valid, human_valid = [], []
    predicted_full, human_full = [], []

    for _, row in dataset.iterrows():
        w1, w2 = row["Word 1"].lower(), row["Word 2"].lower()
        human_score = row["Human (mean)"]

        if w1 in key_index and w2 in key_index:
            sim = 1 - cosine(get_vec(w1), get_vec(w2))
            predicted_valid.append(sim)
            human_valid.append(human_score)
            predicted_full.append(sim)
        else:
            # assign 0 for full version when missing
            predicted_full.append(0)

        human_full.append(human_score)

    # If no valid pairs, set NaN
    if len(predicted_valid) == 0:
        rho_valid, p_valid = np.nan, np.nan
    else:
        rho_valid, p_valid = spearmanr(predicted_valid, human_valid)

    rho_full, p_full = spearmanr(predicted_full, human_full)
    return {
        "rho_valid": rho_valid, "p_valid": p_valid, "n_valid": len(predicted_valid),
        "rho_full": rho_full, "p_full": p_full, "n_full": len(predicted_full)
    }

# Define experiments: you must run at least 2 extra experiments per dataset.
experiments = [
    # small corpus experiments (base + 2 variants)
    {"name": "small_base",     "data": train_small, "params": dict(vector_size=50,  window=5,  min_count=5,  epochs=5)},
    {"name": "small_vec100",   "data": train_small, "params": dict(vector_size=100, window=5,  min_count=5,  epochs=5)},
    {"name": "small_window10", "data": train_small, "params": dict(vector_size=50,  window=10, min_count=5,  epochs=5)},
    # large corpus experiments (base + 2 variants)
    {"name": "large_base",     "data": train_large, "params": dict(vector_size=50,  window=5,  min_count=5,  epochs=5)},
    {"name": "large_vec100",   "data": train_large, "params": dict(vector_size=100, window=5,  min_count=5,  epochs=5)},
    {"name": "large_epoch10",  "data": train_large, "params": dict(vector_size=50,  window=5,  min_count=5,  epochs=10)},
]

results = []

for exp in experiments:
    name = exp["name"]
    data = exp["data"]
    p = exp["params"]
    print(f"\n=== Training {name} | params: {p} ===")
    t0 = time.time()

    # Create model instance (don't pass sentences to constructor if you want to control build_vocab/train)
    model = Word2Vec(vector_size=p["vector_size"], window=p["window"], min_count=p["min_count"], workers=4)

    # build vocab and train explicitly (ensures epochs are used)
    model.build_vocab(data)
    model.train(data, total_examples=model.corpus_count, epochs=p["epochs"])

    t1 = time.time()
    train_time = t1 - t0

    # Evaluate (dual)
    eval_res = evaluate_model_on_wordsim_dual(model, ws353)

    # record stats
    row = {
        "Model": name,
        "corpus": "small" if "small" in name else "large",
        "vector_size": p["vector_size"],
        "window": p["window"],
        "min_count": p["min_count"],
        "epochs": p["epochs"],
        "vocab_size": len(model.wv),
        "train_time_s": round(train_time, 1),
        # valid-only metrics
        "rho_valid": np.round(eval_res["rho_valid"], 4) if not np.isnan(eval_res["rho_valid"]) else eval_res["rho_valid"],
        "p_valid": eval_res["p_valid"],
        "n_valid": eval_res["n_valid"],
        # full (353) metrics
        "rho_full": np.round(eval_res["rho_full"], 4),
        "p_full": eval_res["p_full"],
        "n_full": eval_res["n_full"]
    }
    results.append(row)
    print(f"Done. vocab_size={row['vocab_size']}, rho_valid={row['rho_valid']}, rho_full={row['rho_full']}")

# Convert to dataframe and show
df_results = pd.DataFrame(results)
# order columns nicely
cols = ["Model","corpus","vocab_size","vector_size","window","min_count","epochs","train_time_s",
        "n_valid","rho_valid","p_valid","n_full","rho_full","p_full"]
df_results = df_results[cols]
print("\n--- Experiment Results ---")
display(df_results)
```

**Part III – Word Embeddings Exploration: Analogies. Step 7 – Analogy Task**

```python
# ============================================================
# STEP 7: Word Embeddings Exploration – Analogies
# ============================================================
# References:
# Mikolov, T., Chen, K., Corrado, G., & Dean, J. (2013).
#   Efficient Estimation of Word Representations in Vector Space.
#   arXiv preprint arXiv:1301.3781. (Introduced the parallelogram model for analogies)
#
# Rumelhart, D. E., & Abrahamson, A. A. (1973).
#   A model for analogical reasoning. Cognitive Psychology, 5(1), 1–28.
#   (Early cognitive model inspiring vector analogies)
#
# Bolukbasi, T., Chang, K.-W., Zou, J. Y., Saligrama, V., & Kalai, A. T. (2016).
#   Man is to Computer Programmer as Woman is to Homemaker? Debiasing Word Embeddings.
#   Advances in Neural Information Processing Systems (NeurIPS).
#   (Discusses gender bias in pretrained word embeddings)
#
# Gensim Documentation:
#   https://radimrehurek.com/gensim/models/keyedvectors.html#gensim.models.keyedvectors.KeyedVectors.most_similar
# ============================================================

import gensim.downloader as api

# ------------------------------------------------------------
# 1. Load Pretrained Google News Word2Vec Model
# ------------------------------------------------------------
print("Loading pretrained 'word2vec-google-news-300' model (~1.6GB)...")
model = api.load("word2vec-google-news-300")
print(" Model loaded successfully! Vocabulary size:", len(model))

# ------------------------------------------------------------
# 2. Helper Function for Analogies
# ------------------------------------------------------------
def show_analogies(model, positive_words, negative_words):
    """
    Implements the parallelogram model for analogies:
        'a' is to 'b' as 'c' is to ?
    Formula: vec(b) - vec(a) + vec(c)
    Returns top-5 most similar words.
    """
    try:
        results = model.most_similar(positive=positive_words, negative=negative_words, topn=5)
        print(f"\nAnalogy: {negative_words[0]} is to {positive_words[0]} as {positive_words[1]} is to ___ ?")
        print("Top-5 predictions:")
        for word, score in results:
            print(f"  {word:>15s}  (similarity = {score:.4f})")
        return results
    except KeyError as e:
        print(f"⚠️ Missing word in vocabulary: {e}")
        return None

# ------------------------------------------------------------
# 3. Analogy Tasks (Part g) – Semantic Relations
# ------------------------------------------------------------
print("\n=== Step 7 (g): Analogy Exploration – Semantic Analogies ===")

analogies_g = [
    (["woman", "king"], ["man"]),            # man:woman :: king:?
    (["Greece", "Rome"], ["Athens"]),        # Athens:Greece :: Rome:?
    (["read", "playing"], ["reading"]),      # reading:read :: playing:?
    (["souvlaki", "Italy"], ["Greece"]),     # Greece:souvlaki :: Italy:?
    (["propeller", "car"], ["airplane"])     # airplane:propeller :: car:?
]

results_g = []
for pos, neg in analogies_g:
    res = show_analogies(model, pos, neg)
    results_g.append((pos, neg, res))

# ------------------------------------------------------------
# 4. Analogy Tasks (Part h) – Gender Bias Exploration
# ------------------------------------------------------------
print("\n=== Step 7 (h): Analogy Exploration – Gender Bias Tests ===")

analogies_h = [
    (["woman", "computer_programmer"], ["man"]),
    (["woman", "superstar"], ["man"]),
    (["woman", "guitarist"], ["man"]),
    (["woman", "boss"], ["man"])
]

results_h = []
for pos, neg in analogies_h:
    res = show_analogies(model, pos, neg)
    results_h.append((pos, neg, res))
```

The pretrained Google News Word2Vec model exhibits gender bias, confirming findings by Bolukbasi et al. (2016). These biases likely reflect stereotypes in the massive web-text training data (e.g., associating women with homemaking or support roles).

While analogies like king → queen show legitimate semantic structure, biased analogies highlight the need for debiasing techniques before applying pretrained embeddings in sensitive contexts.
